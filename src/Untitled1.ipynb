```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

% matplotlib inline
```

```python
magic_seaweed_days = ['A', 'B', 'C', 'D', 'E', 'F']
magic_seaweed_H = np.array([4, 3.5, 6, 4, 4, 5]) # these values are in feet
surfline_regional_SF_county_H = np.array([4,3,7,6,7,7]) # these values are in feet
model_days = ['A', 'B', 'C', 'D', 'E', 'F']
model_H = np.array([2.1, 2.0, 1.7, 2.0, 2.0, 2.2]) # these values are in meters
```

```python
fig, ax = plt.subplots(figsize=(18,5))

_= ax.plot(model_days, model_H,'o', 
           label='Model',color = 'r', 
           markersize=12, markeredgecolor='r', 
           markerfacecolor='none')

_= ax.plot(magic_seaweed_days, 
           magic_seaweed_H*0.3,'s', 
           label='Magic seaweed', 
           color='b', markersize=12, 
           markeredgecolor='b', markerfacecolor='none')

_= ax.plot(model_days, 
           surfline_regional_SF_county_H*0.3,'^', 
           label='Surfline - SF Regional',color = 'm', 
           markersize=12, markeredgecolor='m',
           markerfacecolor='none')


_= ax.set_ylim([0,4])
_= ax.set_xticklabels(['Thursday(2/8)', 'Friday(2/9)', 'Saturday(2/10)', 'Sunday(2/11)', 
                       'Monday(2/12)', 'Tuesday(2/13)'], 
                      fontsize=14, 
                      color='k')

_= ax.set_ylabel('Wave Height (m)', fontsize=14, color='k')
_= ax.tick_params(axis='y', labelsize=12, colors='black') 
_= plt.legend()
_=plt.legend(loc=1, prop={'size': 14})
plt.tight_layout()
plt.savefig('Example_Comp_MagicSeaweed_Surfline')
```
